The plot above shows a clear difference in the distribution of adults aged 45-64 between San Francisco and Shasta. #### **Q8.**
1. Use Min-Max Normalization to transform the values of the attribute SMHS1_CT the range 0.0-1.0
2. Perform Z-score Normalization to standardize the values of SMHS1_CT
3. Do both in a new dataframe derived from the original for both Min-Max Normalization and Z-score Normalization for all numeric attributes from the original dataframe in the newly copied and filtered dataframe.

```python
# Create a copy of original dataframe
df_adult_demographics_copy2  = df_adult_demographics.copy()
```

```python
# Generate function for Min-Max Normalization
def min_max_normalization(series):
    return (series - series.min()) / (series.max() - series.min())

# Apply Min-Max Normalization to 'SMHS1_CT'
df_adult_demographics_copy2["MIN_MAX_NORMALIZATION_SMHS1_CT"] = min_max_normalization(df_adult_demographics_copy2["SMHS1_CT"])
df_adult_demographics_copy2["MIN_MAX_NORMALIZATION_SMHS1_CT"].head(5)
```

```python
# Generate function for Z-score Normalization
def z_score_normalization(data_frame):
    return (data_frame - data_frame.mean()) / (data_frame.std())

# Apply Z-score Normalization to 'SMHS1_CT'
df_adult_demographics_copy2["Z_SCORE_NORMALIZATION_SMHS1_CT"] = z_score_normalization(df_adult_demographics_copy2["SMHS1_CT"])
df_adult_demographics_copy2["Z_SCORE_NORMALIZATION_SMHS1_CT"].head(5)
```

```python
# Create a copy of original dataframe
df_adult_demographics_copy3  = df_adult_demographics.copy()
```

```python
# Filter for numeric columns
adult_demographics_numeric_columns = ["TOTAL_CT", "SMHS1_CT", "SMHS5_CT"]

# Create new dataframe with numeric columns
df_adult_demographics_numeric_columns = df_adult_demographics_copy3[adult_demographics_numeric_columns].copy()

# Apply Min-Max Normalization to numeric columns
for column in adult_demographics_numeric_columns:
    df_adult_demographics_numeric_columns[f"MIN_MAX_NORMALIZATION_{column}"] = min_max_normalization(df_adult_demographics_numeric_columns[column])

# Apply Z-score Normalization to numeric columns
for column in adult_demographics_numeric_columns:
    df_adult_demographics_numeric_columns[f"Z_SCORE_NORMALIZATION_{column}"] = z_score_normalization(df_adult_demographics_numeric_columns[column])

# Display the result
df_adult_demographics_numeric_columns.head(5)
```

#### **Q9.**
1. Using Pandas, determine all the attributes with missing values and the number of missing values for each such attribute.
2. Drop the column with over 5700 missing datapoints.
3. Show all the instances in the data that contain a missing value.
4. Fill the missing values for all numeric attributes using the mean value for the attribute.
5. Show that the final resulting table does not contain missing values.

```python
# Calculate the number of missing values in each column
missing_values = df_adult_demographics.isna().sum()

# Display the result
for column, count in missing_values.items():
    if count != 0:
        print(f"The column '{column}' has {count} missing values. \n")
```

```python
# Identify columns with more than 5700 missing values
columns_with_over_5700_missing_values = missing_values[missing_values > 5700].index

# Drop the columns with more than 5700 missing values
df_adult_demographics_cleaned = df_adult_demographics.drop(columns = columns_with_over_5700_missing_values)
df_adult_demographics_cleaned.head(5)
```

```python
# Display all the instances that contain at least one missing value
missing_instances = df_adult_demographics_cleaned[df_adult_demographics_cleaned.isnull().any(axis = 1)]
print(len(missing_instances))
```

```python
# Filter for numeric columns
adult_demographics_numeric_columns = ["TOTAL_CT", "SMHS1_CT", "SMHS5_CT"]

# Fill missing values with the mean of the respective column
for column in adult_demographics_numeric_columns:
    column_mean = df_adult_demographics_cleaned[column].mean()
    df_adult_demographics_cleaned[column] = df_adult_demographics_cleaned[column].fillna(column_mean, axis = 0)

# Display the result
df_adult_demographics_cleaned.head(5)
```

```python
# Display the result
for column in adult_demographics_numeric_columns:
    count = df_adult_demographics_cleaned[column].isna().sum()
    if count != 0:
        print(f"The column '{column}' has {count} missing values. \n")
    else:
        print(f"There is no missing value in '{column}' column. \n")
```
