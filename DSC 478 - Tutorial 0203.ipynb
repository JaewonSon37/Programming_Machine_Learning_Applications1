```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ticker = 'AAPL'
dfdata = yf.download(ticker, start = '2020-01-01', end = '2024-09-12', progress = False)

display(dfdata)
```

```python
dfdata['Return'] = dfdata['Adj Close'].pct_change()
dfdata = dfdata.dropna()
quantiles = pd.qcut(dfdata['Return'], q = 4, labels = False)
dfdata['Quantile'] = quantiles
dfdata
```
